#### 리팩토링 과제

Day4에서 만든 "붕어빵 자동 판매 시스템"은 절차적 코드로 구현되어있습니다.

Day5에서 배운 함수, 클래스 개념을 적용시켜 객체지향적으로 리팩토링을 진행해주세요


```python
# 리팩토링 과제 힌트
# 구조는 아래와 같습니다.
class BungeoppangShop:
    def __init__(self):
        self.붕어빵_재고={"팥붕어빵":10,"슈크림붕어빵":8,"초코붕어빵":5}
        self.가격={"팥붕어빵": 1000,"슈크림붕어빵": 1200,"초코붕어빵": 1500}
        self.판매개수={"팥붕어빵": 0,"슈크림붕어빵": 0,"초코붕어빵": 0}

    def check_stock(self):
        print(f"현재재고=> {self.붕어빵_재고}")

    def process_order(self):
        주문_내역={"팥붕어빵":0,"슈크림붕어빵":0,"초코붕어빵":0}
        주문=input("주문할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 주문을 마치시려면 '종료' 입력")
        while 주문!="종료":
            붕어빵_종류,붕어빵_개수=주문.split(" ")
            주문_내역[붕어빵_종류]+=int(붕어빵_개수)
            주문=input("주문할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 주문을 마치시려면 '종료' 입력")
        for k,v in 주문_내역.items():
            if self.붕어빵_재고[k]<v:
                print("재고가 부족합니다!")
            elif v!=0:
                self.붕어빵_재고[k]-=v
                self.판매개수[k]+=v
        self.check_stock()

    def admin_mode(self):
        관리자_입력=input("추가할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 메인화면으로 가실려면 '종료' 입력")
        while 관리자_입력!="종료":
            붕어빵_종류,붕어빵_개수=관리자_입력.split(" ")
            self.붕어빵_재고[붕어빵_종류]+=int(붕어빵_개수)
            관리자_입력=input("추가할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 메인화면으로 가실려면 '종료' 입력")
        self.check_stock()

    def open(self):
        관리자_입력=input("숫자를 입력하세요. 1.관리자_모드  2.주문  (장사를 끝마치시려면 '종료' 입력)")
        while 관리자_입력!="종료":
            if 관리자_입력=="1":
                self.admin_mode()
            elif 관리자_입력=="2":
                self.process_order()
            else:
                print("1 또는 2를 입력해주세요")
            관리자_입력=input("숫자를 입력하세요. 1.관리자_모드  2.주문  (장사를 끝마치시려면 '종료' 입력)")
        print("장사가 종료되었습니다.")
        print(f"현재 판매개수=> {self.판매개수}")
        print(f"현재재고=> {self.붕어빵_재고}")
        self.calculate_total_sales()

    def calculate_total_sales(self):
        총매출=0
        for i in self.붕어빵_재고.keys():
            총매출+=self.가격[i]*self.판매개수[i]
        print(f'총매출 : {총매출}')
        
def main():
    shop=BungeoppangShop()
    shop.open()

if __name__ == "__main__":
    main()
```

- check_stock: 현재 재고를 출력합니다.
- process_order: 주문을 처리하고 재고를 업데이트합니다.
