```python
주문_내역={"팥붕어빵":0,"슈크림붕어빵":0,"초코붕어빵":0}
주문=input("주문할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별)")
while 주문!="종료":
    붕어빵_종류,붕어빵_개수=주문.split(" ")
    주문_내역[붕어빵_종류]+=int(붕어빵_개수)
    print(주문_내역)
    주문=input("주문할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별)")
for k,v in 주문_내역.items():
    if 붕어빵_재고[k]<v:
        print("재고가 부족합니다!")
    elif v!=0:
        붕어빵_재고[k]-=v
        print(f'판매완료 => {k} : {v}')
print(f"현재재고=> {붕어빵_재고}")
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
붕어빵_재고={"팥붕어빵":10,"슈크림붕어빵":8,"초코붕어빵":5}

가격={"팥붕어빵": 1000,
     "슈크림붕어빵": 1200,
     "초코붕어빵": 1500}

판매개수={"팥붕어빵": 0
    ,"슈크림붕어빵": 0
    ,"초코붕어빵": 0}

def 주문(붕어빵_재고,판매개수):
    주문_내역={"팥붕어빵":0,"슈크림붕어빵":0,"초코붕어빵":0}
    주문=input("주문할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 주문을 마치시려면 '종료' 입력")
    while 주문!="종료":
        붕어빵_종류,붕어빵_개수=주문.split(" ")
        주문_내역[붕어빵_종류]+=int(붕어빵_개수)
        주문=input("주문할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 주문을 마치시려면 '종료' 입력")
    for k,v in 주문_내역.items():
        if 붕어빵_재고[k]<v:
            print("재고가 부족합니다!")
        elif v!=0:
            붕어빵_재고[k]-=v
            판매개수[k]+=v
    print(f"현재재고=> {붕어빵_재고}")

def 관리자모드(붕어빵_재고):
    관리자_입력=input("추가할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 메인화면으로 가실려면 '종료' 입력")
    while 관리자_입력!="종료":
        붕어빵_종류,붕어빵_개수=관리자_입력.split(" ")
        붕어빵_재고[붕어빵_종류]+=int(붕어빵_개수)
        관리자_입력=input("추가할 붕어빵의 종류와 개수를 입력하세요(띄어쓰기로 구별) 메인화면으로 가실려면 '종료' 입력")
    print(f"현재재고=> {붕어빵_재고}")


관리자_입력=input("숫자를 입력하세요. 1.관리자_모드  2.주문  (장사를 끝마치시려면 '종료' 입력)")
while 관리자_입력!="종료":
    if 관리자_입력=="1":
        관리자모드(붕어빵_재고)
    elif 관리자_입력=="2":
        주문(붕어빵_재고,판매개수)
    else:
        print("1 또는 2를 입력해주세요")
    관리자_입력=input("숫자를 입력하세요. 1.관리자_모드  2.주문  (장사를 끝마치시려면 '종료' 입력)")

print("장사가 종료되었습니다.")
print(f"현재 판매개수=> {판매개수}")
print(f"현재재고=> {붕어빵_재고}")

총매출=0
for i in 붕어빵_재고.keys():
    총매출+=가격[i]*판매개수[i]
print(f'총매출 : {총매출}')
```
