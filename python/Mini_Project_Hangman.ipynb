#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

10번만에 맞히지 못하면 게임이 끝납니다. 🥲



#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.   
문제가 어려우신 경우 구글링 통해 문제를 해결해보세요.  

```python
# 기본 라이브러리라 파이썬 설치시 자동으로 설치됩니다!
# random 모듈은 랜덤한 숫자를 생성할 때 사용합니다.
import random
import os

# 게임에 사용될 단어 목록
words = ["apple", "banana", "orange", "grape", "lemon"]

# 행맨 그림
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---""",
]


class HangmanGame:
    def __init__(self,words):
        self.count=7
        self.answer=words[random.randrange(0,5)]
        self.user_answer=["_"]*len(self.answer)
        pass

    def display(self):
        print(hangman_pics[7-self.count])
        pass
    def word_correct(self):
        alpha=input("알파벳 하나를 입력해주세요")
        if alpha in self.answer:
            if alpha in self.user_answer:
                print("이미 맞춘 알파벳입니다! 다시 입력해 주세요.")
                return True
            idx=[i for i,v in enumerate(self.answer) if v==alpha]
            for i in idx:
                self.user_answer=list(self.user_answer)
                self.user_answer[i]=alpha
            if '_' in self.user_answer:
                os.system('cls')
                self.display()
                print("단어를 맞추셨군요. 더 힘내봐요")
            else:
                os.system('cls')
                print(str(self.user_answer))
                self.count=7
                self.answer=words[random.randrange(0,5)]
                self.user_answer=["_"]*len(self.answer)
                return True if 'y'==input("정답을 맞추셨습니다! 정말 대단하네요. 다시 하시겠습니까? y/n") else False
            
        elif self.count!=1:
            self.count-=1
            os.system('cls')
            self.display()
            print(f"틀렸습니다. 남은 시도 횟수 {self.count}")
        else:
            os.system('cls')
            self.count=7
            self.answer=words[random.randrange(0,5)]
            self.user_answer=["_"]*len(self.answer)
            return True if 'y'==input("남은 시도횟수를 다 소모하셨습니다. 다시 하시겠습니까? y/n") else False
        print(str(self.user_answer))
        return True
            

    def play(self):
        while self.word_correct():
            pass

if __name__ == "__main__":
    game = HangmanGame(words)
    game.word_correct()
    game.play()
```
